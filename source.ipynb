Load Datasets

```python
import pandas as pd
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
df_dev=pd.read_csv("RSD_Particles/development.csv")
df_eval=pd.read_csv("RSD_Particles/evaluation.csv",index_col=0)
display(df_dev)
display(df_eval)
```

```python
coordinates=df_dev[['x','y']]
features=df_dev.drop(['x','y'],axis=1)
display(coordinates)
display(features)

evaluation_data=df_eval
```

```python
#Checking NaN features
pd.set_option('display.max_rows', None)
print(features.isna().any(axis=0))
pd.reset_option('display.max_rows')
```

```python
#scatter
plt.scatter(coordinates['x'], coordinates['y'])
plt.xlabel('Coordinate X')
plt.ylabel('Coordinate Y')
plt.title('Scatter Plot of X e Y')
plt.show()
```

```python
#means
fig, axes = plt.subplots(1, 5, figsize=(20, 4), sharey=True)
for i, feature_type in enumerate(['pmax', 'negpmax', 'tmax', 'rms', 'area']):
    mean_values = features[[f'{feature_type}[{j}]' for j in range(18)]].mean()
    axes[i].bar(range(1, 19), mean_values)
    axes[i].set_title(f'Mean for {feature_type}')
    axes[i].set_xlabel('Feature Index')
    axes[i].set_ylabel('Mean')
    axes[i].grid(visible=True, which='both')
    axes[i].set_xticks(range(1, 19))
fig.text(0.5, 0, 'Mean', ha='center', va='center')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
#removing features after means
columns_to_drop = ['tmax[0]', 'tmax[1]', 'tmax[2]', 'tmax[3]', 'tmax[4]', 'tmax[5]', 'tmax[6]', 'tmax[7]', 'tmax[8]', 'tmax[9]', 'tmax[10]', 'tmax[11]', 'tmax[12]', 'tmax[13]', 'tmax[14]', 'pmax[15]', 'area[15]', 'tmax[15]', 'pmax[16]', 'area[16]', 'tmax[16]', 'pmax[17]', 'area[17]', 'tmax[17]']
columns_to_drop = [col for col in columns_to_drop if col in features.columns]
features = features.drop(columns_to_drop, axis=1, errors='ignore')
sorted_features = features.sort_index(axis=1)
evaluation_data=evaluation_data.drop(columns=columns_to_drop)
display(evaluation_data)
display(features)
```

```python
#dev standard
pd.set_option('display.max_rows', None)
std = sorted_features.std()
pd.reset_option('display.max_rows')
plt.figure(figsize=(15, 8))
plt.bar(sorted_features.columns, std)
plt.xticks(rotation=90)
plt.title('Standard Deviation for Feature')
plt.show()
```

```python
#removing rms after std
columns_to_drop = [f'rms[{i}]' for i in range(18)]  # Use a list comprehension to generate column names
columns_to_drop = [col for col in columns_to_drop if col in features.columns]
evaluation_data=evaluation_data.drop(columns=columns_to_drop)
features = features.drop(columns_to_drop, axis=1, errors='ignore')
sorted_features = features.sort_index(axis=1)
display(evaluation_data)
display(features)
```

```python
#boxplot
import seaborn as sns
macrofeatures = ['pmax', 'negpmax', 'area']
for macrofeature in macrofeatures:
    plt.figure(figsize=(8, 6))
    columns_of_interest = [f"{macrofeature}[{i}]" for i in range(18) if f"{macrofeature}[{i}]" in features.columns]
    data_to_plot = features[columns_of_interest]
    ax = sns.boxplot(data=data_to_plot)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
    plt.title(f'Boxplot for {macrofeature}')
    plt.xlabel(macrofeature)
    plt.ylabel('Values')
    plt.show()

display(evaluation_data)
display(features)
```

```python
#removing features after boxplot
columns_to_drop = ['pmax[0]', 'pmax[7]', 'pmax[12]']
columns_to_drop = [col for col in columns_to_drop if col in features.columns]
evaluation_data=evaluation_data.drop(columns=columns_to_drop)
features = features.drop(columns_to_drop, axis=1, errors='ignore')
sorted_features = features.sort_index(axis=1)
display(evaluation_data)
display(features)
```

```python
#correlation analysis
correlation_matrix = features.corr()
plt.figure(figsize=(50, 50))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5)
plt.title('Features Correlation Matrix')
plt.show()
```

```python
#removing features after corr analysis
columns_to_drop = ['area[0]', 'area[1]', 'area[6]', 'area[7]', 'area[9]', 'area[12]','area[14]']
columns_to_drop = [col for col in columns_to_drop if col in features.columns]
evaluation_data=evaluation_data.drop(columns=columns_to_drop)
features = features.drop(columns_to_drop, axis=1, errors='ignore')
sorted_features = features.sort_index(axis=1)
display(evaluation_data)
display(features)
```

GridSearchCV(Random Forest Regressor)

```python
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
from sklearn.metrics.pairwise import paired_distances
from sklearn.ensemble import RandomForestRegressor

def euclidean_distance(y_true, y_pred):
    return np.sum(paired_distances(y_true, y_pred))/len(y_true)

euclidean_scorer = make_scorer(euclidean_distance, greater_is_better=False)

#grid
param_grid = {
    'n_estimators': [100, 200, 250],
    'max_features':['sqrt','log2'],
    'random_state':[42],
    'criterion': ['squared_error'],
    'n_jobs':[-1] 
}
grid_search = GridSearchCV(RandomForestRegressor(), param_grid, scoring=euclidean_scorer, n_jobs=-1, cv=5)
grid_search.fit(X_train, y_train)
print("Best Hyperparameters:", grid_search.best_params_)
best_rf=grid_search.best_estimator_
print(best_rf)
predictions_val_rf = best_rf.predict(X_val)
euclidean_distances_val_rf = euclidean_distance(y_true=y_val, y_pred=predictions_val_rf)
average_euclidean_distance_val_rf = np.mean(euclidean_distances_val_rf)
print(f"Avg euclidean distance on validation (RF): {average_euclidean_distance_val_rf}")

#predict
predictions_eval_rf = best_rf.predict(evaluation_data)
print(grid_search.best_params_)
predictions_df = pd.DataFrame({
    'Id': evaluation_data.index,
    'Predicted': [f"{x:.1f}|{y:.1f}" for x, y in predictions_eval_rf]
})

# csv
predictions_df.to_csv('DSL_Winter_Project_2024/predictions.csv', index=False)
```
